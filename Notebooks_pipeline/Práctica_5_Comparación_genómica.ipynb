## **Practica 5. Comparación de genomas bacterianos *(Salmonella)***

Este instructivo tiene como objetivo que cada participante realice un análisis de genómica comparativa de cepas de *Salmonella* a partir de datos de secuenciación.

Para ello, se utilizará la herramienta **Roary** con el fin de obtener el pangenoma de las cepas analizadas. Además, se empleará **SNP-sites** para extraer los SNPs a partir del alineamiento del genoma central.

Posteriormente, se utilizará **IQ-TREE2** para construir el árbol filogenético basado en los SNPs identificados y **SNP-dists** para calcular la matriz de distancias de SNPs entre las muestras.

## **¿Qué hacer con los archivos Fastq tras secuenciar una muestra de Salmonella?**

El análisis bioinformático de datos de secuenciación generalmente incluye los siguientes pasos:

1. **Control de calidad (QC)**: FastQC, TrimGalore, Kraken2.

2. **Ensamblaje de genomas**: Unicycler.

3. **Caracterización y tipificación**: identificación de determinantes de resistencia mediante SeqSero2 y AMRFinderPlus.

4. **Análisis filogenético**: **construcción de relaciones evolutivas entre las muestras**.

Este instructivo se centra específicamente en el uso de **Roary**, **SNP-sites**, **IQ-TREE2** y **SNP-dists** para realizar análisis de genómica comparativa.

Cree la carpeta **5-Practica_Comparacion**, donde se almacenarán los resultados obtenidos en esta etapa del pipeline, y luego ingrese a ella:

```
cd 5-Practica_Comparacion
```

Para esta práctica se trabajará con un [caso de estudio](https://github.com/mecalderon/Tutorial_Panama/blob/main/Caso_Estudio/Caso_Genómica_Comparativa.pdf). Por ello, se creará un acceso directo a las secuencias:
```
mkdir fasta
ln -s /home/jovyan/work/Tutorial_Panama/Caso_Estudio/FASTA/*.fasta fasta
```

Active el ambiente necesario para ejecutar las herramientas:
```
source activate microenv
```

Inicialmente, cada participante anotará un único genoma utilizando la herramienta **Prokka**. Ejecute el siguiente comando para generar los archivos GFF, reemplazando **PANXXX** por el nombre del genoma correspondiente:
```
prokka --prefix PANXXX --outdir 1-prokka/PANXXX --addgenes --protein /home/jovyan/databases/microanalysis/prokka/salmonella_enterica.faa --strain PANXXX --mincontiglen 300 --cpus 4 fasta/PANXXX.fasta
```

Para obtener el pangenoma de las cepas a analizar, se utilizarán las anotaciones de todas las muestras. Use el siguiente comando para crear un acceso directo a dichas anotaciones:
```
ln -s /home/jovyan/work/Tutorial_Panama/Caso_Estudio/Prokka/*.gff 1-prokka
```

El alineamiento del genoma core se realizará con la herramienta **Roary**. El siguiente comando generará el core genome en la carpeta de salida llamada **2-roary**:
```
roary -e --mafft -f 2-roary -p 4 1-prokka/*.gff
```
Dentro de la carpeta **2-roary** se encuentra el archivo **core_gene_alignment.aln**, que corresponde al alineamiento del genoma core. Para obtener los SNPs a partir de este alineamiento, se utilizará la herramienta **SNP-sites**.

Primero, cree la carpeta **3-snpsites** para almacenar el alineamiento de SNPs:
```
mkdir 3-snpsites
```

Ejecute el siguiente comando para extraer los SNPs del alineamiento:
```
snp-sites -o 3-snpsites/core_gene_alignment_snps.aln 2-roary/core_gene_alignment.aln
```

A continuación, se construirá el árbol filogenético basado en los SNPs identificados utilizando la herramienta **IQ-TREE2**. Para ello, cree el directorio **4-iqtree2** e ingrese en él:
```
mkdir 4-iqtree2
cd 4-iqtree2
```

Ejecute el siguiente comando para generar el árbol filogenético con **IQ-TREE2**:
```
iqtree2 -s /home/jovyan/work/5-Practica_Comparacion/3-snpsites/core_gene_alignment_snps.aln -m GTR+ASC -pre Salmonella_roary -bb 10000 -nt 4
```
Dentro de la carpeta **4-iqtree2** se encuentra el árbol filogenético llamado **Salmonella_roary.treefile**. Descárguelo a su computadora local y visualícelo en [iTOL](https://itol.embl.de/).

Para generar la matriz de distancias entre las muestras, se utilizará la herramienta **SNP-dists**. Primero, cree la carpeta **5-snpdist**, donde se almacenará la matriz de SNPs derivada del alineamiento:
```
cd ..
mkdir 5-snpdist
```

Ejecute el siguiente comando para generar la matriz de distancias:
```
snp-dists 3-snpsites/core_gene_alignment_snps.aln > 5-snpdist/roary_snp_matrix.tsv
```
Dentro de la carpeta **5-snpdist** se encontrará la matriz generada. Descárguela y proceda a su análisis.
